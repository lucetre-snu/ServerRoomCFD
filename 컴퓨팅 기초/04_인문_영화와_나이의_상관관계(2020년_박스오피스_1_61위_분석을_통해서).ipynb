**(각 조원이 맡았던 코드1은 전체 형식은 통일했지만, 세부적 부분에서는 개성을 살리기로 했습니다. 이에 6개의 코드1은 변수명에 있어 약간의 차이를 보일 수 있음을 감안해주시기 바랍니다. 구현방법에 대한 설명은 2020 박스오피스 21-30위 영화 코드를 기준으로 작성했습니다.)**
*   **설명** :   해당 코드는 영화별 ‘나이대’와 ‘나이대별 관람 비율’에 대한 정보를 수집하는 코드입니다. 
Chrome 개발자 도구와 BeautifulSoup 모듈을 이용해 원하는 정보를 Wepscrapping 한 것이 특징입니다. 

  2020년 박스오피스 1-61위의 영화를 조원 한 명당 10 작품씩 맡아 Webscrapping 했습니다. (이때 60위의 작품에 대한 정보가 나오지 않아 61위를 조사한 것입니다.) 영화의 나이대별 관람 비율에 대해 사이트에서 제공하는 정보는 10대-50대로 한정되어 있었기에, 저희 조에서도 각 영화의 10대-50대의 관람 비율만을 스크랩하기로 했습니다. 

  저희가 스크랩한 웹사이트는 네이버 검색창에 ‘영화 제목’을 검색했을 때의 사이트이고, 이때의 url을 이용한 것입니다. 각 url에서 개발자도구를 이용해 확인한 결과, 영화제목에 해당하는 tag명과 class명은 각각 'strong', '_text'로 동일했고, 나이대는 ‘span’, ‘area_text_age’, 나이대별 관람 비율은 'div','area_text_percent'의 tag명과 class명을 동일하게 갖고 있었습니다.

(50~61위의 경우, 한 영화에 대한 관람 비율 데이터가 부재하여 61위의 영화를 추가하였습니다.)

 각각의 순위별 분석에서 모두 전반적인 추세는 20대, 30대, 40대, 50대, 그리고 10대 순으로 관람 비율이 높게 나타남을 도출할 수 있었습니다. 다만, 51~61위의 산점도 그래프의 분포 양상이 특징적인데, 40대와 50대의 관람 비율도 비교적 높게 나타나고 있습니다. 이는 50대 관람 비율이 높은 <미스터트롯: 더 무비>이나, 40대의 관람 비율이 비교적 높은 <애니멀 크래커>, <수퍼 소닉> 등에 의한 예외적인 사례들에 기인했다고 해석할 수 있습니다. 결국 순위대별로 분할했음에도, 전반적인 관람 비율 추세가 동일하다는 점은 전술한 예외들을 제외하고 거의 모든 장르의 영화에서 공통적으로 이 결과가 적용됨을 시사한다고 볼 수 있습니다. 

 이후 1위~61위의 데이터를 바탕으로 한 산점도를 제작하여 종합적인 결론을 도출한 결과, 각각의 순위대별을 분산하여 얻은 결론과 유사함을 확인할 수 있었습니다. 20대, 30대, 40대, 50대, 그리고 10대 순으로 높게 나타난 것입니다. 이를 통해 '10대를 제외하고 나이가 적을수록 영화관에서 영화를 많이 본다'는 결론을 도출할 수 있었습니다. (10대의 경우 경제적, 학업적 요인 등에 의해 가장 낮게 나타난 것으로 추정해보았습니다. )
