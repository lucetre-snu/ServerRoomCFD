# #1 Matrix Calculator

```python
def str_to_mat(s):
    M = s.split('],[')
    M[0] = M[0][2:]
    M[-1] = M[-1][:-2]
    return [[int(e) for e in row.split(',')] for row in M]

def mat_len(M):
    return (len(M), len(M[0]))

def mat_to_str(M):
    return '[[' + '],['.join([','.join([str(e) for e in row]) for row in M]) + ']]'

def minusM(A):
    return [[-e for e in row] for row in A]

def tranM(A):
    return [[e for e in row] for row in zip(*A)]

def sumM(A, B):
    if mat_len(A) != mat_len(B):
        raise Exception()
    return [[sum(pair) for pair in zip(*rows)] for rows in zip(A, B)]

def subM(A, B):
    if mat_len(A) != mat_len(B):
        raise Exception()
    return [[sum(pair) for pair in zip(*rows)] for rows in zip(A, minusM(B))]

def mulM(A, B):
    if mat_len(A)[1] != mat_len(B)[0]:
        raise Exception()
    return [[sum(a*b for a,b in zip(rowA, colB)) for colB in zip(*B)] for rowA in A]

A = str_to_mat(input())
B = str_to_mat(input())

print(mat_to_str(minusM(A)))
print(mat_to_str(tranM(A)))
    
try: print(mat_to_str(sumM(A, B)))
except: print('Error')
    
try: print(mat_to_str(subM(A, B)))
except: print('Error')
    
try: print(mat_to_str(mulM(A, B)))
except: print('Error')
```

# #2 Square Roots via Newtons Method (Revisit)

```python
from fractions import Fraction
from math import sqrt

a = Fraction(int(input()))
x = Fraction(int(input()))
eps = float(input())
f = Fraction(sqrt(a)).limit_denominator(x.denominator)

while (x*x-a)/a <= -eps or (x*x-a)/a >= eps:
    print(x, f)
    x = Fraction(1/2) * (x + a/x)
    f = Fraction(sqrt(a)).limit_denominator(x.denominator)
print(x, f)
```

```python
from fractions import Fraction
from math import sqrt

for i in range(10):
    tn = "%02d" % (i+1)
    with open(f'./sqrt/in{tn}.txt', 'r') as rf:
        with open(f'./sqrt/out{tn}.txt', 'w') as wf:

            a_str = rf.readline().rstrip('\n')
            x_str = rf.readline().rstrip('\n')
            eps_str = rf.readline().rstrip('\n')

            a = Fraction(int(a_str))
            x = Fraction(int(x_str))
            eps = float(eps_str)
            f = Fraction(sqrt(a)).limit_denominator(x.denominator)

            while (x*x-a)/a <= -eps or (x*x-a)/a >= eps:
#                 print(x, f)
                wf.write(str(x)+' '+str(f)+'\n')
                x = Fraction(1/2) * (x + a/x)
                f = Fraction(sqrt(a)).limit_denominator(x.denominator)
#             print(x, f)
            wf.write(str(x)+' '+str(f)+'\n')
```

# #3 Ant Experiment

```python
from decimal import *
pos_list = input().split()
dir_list = input().split()

ants = []
for pos, dir in zip(pos_list, dir_list):
    if dir == '+':
        ants.append(1-Decimal(pos))
    else: ants.append(Decimal(pos))
print(format(max(ants), '.20f'))
print(format(min(ants), '.20f'))
```

```python
import numpy as np
for i in range(10):
    tn = "%02d" % (i+1)
    with open(f'./ant/in{tn}.txt', 'w') as wf:
        n = np.random.randint(1, 100)
        for i in np.random.rand(n):
            m = np.random.randint(1, 20)
            print(round(i, m), end=' ')
            wf.write(str(round(i, m)) + ' ')
        print()
        wf.write('\n')
        for i in np.random.rand(n):
            m = np.random.randint(2)
            if m == 0:
                print('+', end=' ')
                wf.write('+ ')
            else:
                print('-', end=' ')
                wf.write('- ')
        wf.write('\n')
```
